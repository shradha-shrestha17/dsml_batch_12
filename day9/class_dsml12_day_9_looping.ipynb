```python
#break

count = 1
while count < 10:
  if count ==4:
    break
  print (count)
  count += 1
print("Loop Ended")
```

```python
#continue

count = 0
while count < 10:
  count +=1
  if count ==3:
    continue
  print (count)
print("Loop Ended")
```

```python
#pass

count = 0
while count < 10:
  count +=1
  if count ==3:
    pass
  print (count)
print("Loop Ended")
```

```python
balance = 5000

while True:
  print("\n ----ATM MENU----")
  print("1. Check Balance")
  print("2. Deposit")
  print("3. Exit")

  choice = int(input("choose option"))

  #invalid choice
  if choice < 1 or choice >3:
    print("invalid choice")
    continue

  #check balance
  if choice == 1:
    print("Balance is:", balance)

  #Deposit money
  elif choice ==2:
    amount = int(input("Enter deposit amout:"))
    balance += amount
    print("updated balance:",balance)

  #exit
  else:
    print("thank you")
    break
  # elif choice==3:
  #   print("Thank you for using ATM")
  #   break
```
