write a function total_salary(*args) that accepts an employee's salary components (basic salary, allowance, bonus etc) and returns total salary


```python
def total_salary(*args):
  total = 0
  for salary in args:
    total += salary
  return total
```

```python
result = total_salary(15000, 5000, 2000)
print("Total salary:" , result)
```

```python
#another way
def total_salary(*args):
  return sum(args)
```

```python
total_salary(2000, 2000, 2000)
```
